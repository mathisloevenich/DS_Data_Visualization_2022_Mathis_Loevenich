```python
from urllib import request
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
url = "https://docs.google.com/spreadsheets/d/1ESnyJ4b7m96OCjs3GSQ6EGF7YOMuv0XV-ROXTYIazTs/export?format=xlsx"
```

```python
tesla_deaths = pd.read_excel(url)
```

```python
tesla_deaths.info()
```

```python
# get only relevant columns
rel_data = tesla_deaths[["Year", "Date", "Country", "State", "Description", "Deaths", "Model", "AutoPilot claimed", "Verified Tesla Autopilot Death"]]
```

```python
# ----- Preprocessing ------
# remove empyt rows with no date (excel file usually has a summation at the end which is now a valid df row -> needs to be removed) 
rel_data = rel_data[rel_data["Date"].notna()]
rel_data["Year"] = rel_data["Year"].apply(int)
# turn "-" character into Nan
rel_data[rel_data == "-"] = np.nan
rel_data["Deaths"] = rel_data["Deaths"].astype(int)
rel_data["Verified Tesla Autopilot Death"] = pd.to_numeric(rel_data["Verified Tesla Autopilot Death"]).fillna(0)
rel_data["AutoPilot claimed"] = pd.to_numeric(rel_data["AutoPilot claimed"]).fillna(0)
```

```python
autosteer = pd.read_excel("tesla.xlsx", sheet_name="PE16_007_PRODUCTION DATA")
```

```python
autosteer = autosteer.dropna(how='all', axis=0)
```

```python
autosteer_rel  = autosteer[
    ["1l Previous Mileage before Autosteer Install", 
     "1l Next Mileage after Autosteer Install", 
     "Miles before Autosteer", 
     "Miles after Autosteer", 
     "Airbag events before Autosteer", 
     "Airbag events after Autosteer"]
][:43781] # Raster geht bis 43782]
```

```python
autosteer_rel
```

```python
# The only data where you can be sure when the autopilot was installed is when
# data["1l Previous Mileage before Autosteer Install"] == data[""1l Next Mileage after Autosteer Install""]
data_of_interest = autosteer_rel[
    autosteer_rel["1l Previous Mileage before Autosteer Install"] == autosteer_rel["1l Next Mileage after Autosteer Install"]
]
```

```python
data_of_interest
```

```python
sums = data_of_interest.sum()
print(sums)
airbag_per_miles_before = sums["Airbag events before Autosteer"]*1000000/sums["Miles before Autosteer"]
airbag_per_miles_after = sums["Airbag events after Autosteer"]*1000000/sums["Miles after Autosteer"]
print("Before: ", airbag_per_miles_before)
print("After: ", airbag_per_miles_after)
```

```python
# Now look at data where autosteer actually got installed after driving the vehicle for a while
mileage_greater_zero = data_of_interest[data_of_interest["Miles before Autosteer"] > 0]
```

```python
mileage_greater_zero
```

```python
# Some values are negative (probably an error) and we can't interprete these
mileage_greater_zero = mileage_greater_zero[mileage_greater_zero["Miles after Autosteer"]>0]
```

```python
sums = mileage_greater_zero.sum()
print(sums)
airbag_per_miles_before = sums["Airbag events before Autosteer"]*1000000/sums["Miles before Autosteer"]
airbag_per_miles_after = sums["Airbag events after Autosteer"]*1000000/sums["Miles after Autosteer"]
print("Before: ", airbag_per_miles_before)
print("After: ", airbag_per_miles_after)
```
